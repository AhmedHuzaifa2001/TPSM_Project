```python
import pandas as pd

df = pd.read_csv("dataset.csv")
df.head()
```

```python
## Check for duplicate tracks

duplicate_tracks = df['track_id'].duplicated().sum()
duplicate_tracks
```

```python
df_before = len(df)
df_before
```

```python
# Remove duplicate track_ids, keeping the first occurrence

df = df.drop_duplicates(subset='track_id', keep='first')
df_after = len(df)

print(f"Rows before: {df_before:,}")
print(f"Rows after: {df_after:,}")
print(f"Duplicates removed: {df_before - df_after:,}")
print(f"\n✓ Clean dataset now has {df_after:,} unique tracks")
```

```python
# Create super genre groupings (your cleaner format!)
genre_mapping = {
    "Metal": [
        "black-metal", "death-metal", "heavy-metal",
        "metal", "metalcore"
    ],
    "Rock": [
        "rock", "alt-rock", "hard-rock", "rock-n-roll",
        "psych-rock", "punk-rock", "indie-rock", "garage",
        "grunge", "progressive-rock"
    ],
    "Pop": [
        "pop", "pop-film", "power-pop", "dance",
        "disco", "synth-pop", "dancehall", "k-pop",
        "j-pop", "indie-pop"
    ],
    "Electronic": [
        "edm", "electro", "electronic",
        "techno", "house", "trance",
        "dubstep", "drum-and-bass", "hardstyle",
        "deep-house", "minimal-techno"
    ],
    "Hip-Hop": [
        "hip-hop", "rap", "trap", "trip-hop",
        "grindcore", "breakbeat"
    ],
    "R&B/Soul": [
        "r-n-b", "soul", "funk", "gospel"
    ],
    "Jazz/Blues": [
        "jazz", "blues", "bluegrass", "swing"
    ],
    "Classical": [
        "classical", "opera", "piano"
    ],
    "Acoustic": [
        "acoustic", "folk", "singer-songwriter",
        "indie", "songwriter"
    ],
    "Latin": [
        "latino", "latin", "salsa", "samba",
        "tango", "reggaeton", "spanish", "brazil",
        "MPB", "forro", "pagode", "sertanejo"
    ],
    "Country": [
        "country", "honky-tonk"
    ],
    "World": [
        "reggae", "ska", "world-music", "indian",
        "afrobeat", "iranian", "turkish",
        "malay", "mandopop", "cantopop", "french",
        "swedish", "german", "british"
    ],
    "Ambient": [
        "ambient", "chill", "sleep", "study",
        "new-age", "industrial"
    ],
    "Alternative": [
        "alternative", "emo", "goth",
        "punk", "hardcore",
        "postpunk", "screamo"
    ],
    "Other": [
        "children", "kids", "comedy", "anime",
        "disney", "movies", "show-tunes", "romance",
        "sad", "happy", "party", "work-out",
        "summer", "road-trip"
    ]
}

```

```python
genre_map_flat = {}
for super_genre, genres in genre_mapping.items():
    for genre in genres:
        genre_map_flat[genre] = super_genre

# Apply the mapping
df['super_genre'] = df['track_genre'].map(genre_map_flat)

# For any genres not in our mapping, label them as 'Other'
df['super_genre'] = df['super_genre'].fillna('Other')

# Check the results
print("✓ Super Genre created!")
print(f"\nSuper Genre distribution:")
print(df['super_genre'].value_counts())

print(f"\nTotal Super Genres: {df['super_genre'].nunique()}")
```
